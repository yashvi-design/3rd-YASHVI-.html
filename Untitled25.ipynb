```python
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1


        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def printArray(arr):
    for i in range(len(arr)):
        print(arr[i], end=" ")
    print()

if __name__ == "__main__":
    arr = [12, 11, 13, 5, 6]
    insertionSort(arr)
    printArray(arr)
```

```python
def OnTime(n):
  iteration=0
  for i in range(1,n+1):
    iteration+=1
  print("When n is ",n," Iterations =",iteration)

OnTime(10)
OnTime(20)
OnTime(42)

print("\nWith every 'n' the time taken and iterations will increase")
```

```python
def ONSquareTime(n):
  iteration=0
  for i in range(0,n):
    for j in range(0,n):
      print("*", end=" ")
      iteration+=1
    print("")
  print("\nWhen n is ",n," Iterations = ",iteration,"\n")

ONSquareTime(5)
ONSquareTime(4)
ONSquareTime(3)

print("\nWith every 'n' the time taken equals n^2")
print("0(n^2)")
```
