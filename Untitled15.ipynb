```python
from re import L
lst =['Apple', 'Banana', 'Mango', 'kiwi', 'Blueberry']

print("Length of list", len(lst))
print("First Element:", lst[0])
print("Last letter:", lst[-1])

lst.append('Blackberry')
print("Updateed List :", lst)

lst.remove('Banana')
print("Updated List :", lst)

lst.sort()
print("Sorted list:", lst)

lst.pop(1)
print("Updated List :", lst)

lst.reverse()
print("Reversed List :", lst)

print("Multiplication on list :", lst*2)

lst = lst[:4]
print("Sliced List :", lst)

lst.clear()
print("Updated List :", lst)
```
