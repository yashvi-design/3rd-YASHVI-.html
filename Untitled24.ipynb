```python
def fun1(n):

    return n*(n+1)/2

print(fun1(4))
```

```python
def fun2(n):

    sum=0

    for i in range(1,n+1):

        sum+=i

    return sum

print(fun2(4))
```

```python
def fun3(n):

    sum=0

    for i in range(1,n+1):

        for j in range(1,i+1):

            sum+=1

    return sum

print(fun3(4))
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:

               arr[j], arr[j + 1] = arr[j + 1], arr[j]
```
